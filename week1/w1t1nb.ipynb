```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
page = requests.get("https://en.wikipedia.org/wiki/List_of_languages_by_total_number_of_speakers")

soup = BeautifulSoup(page.content, "html.parser")
```

```python
print(soup.prettify())
```

```python
table = soup.find('table', class_ = "static-row-numbers sortable wikitable")
print(table)
```

```python
titles = table.find_all('th')
```

```python
headers = [title.text.strip() for title in titles]
print(headers)
```

```python
df = pd.DataFrame(columns = headers)
df
```

```python
column_data = table.find_all('tr')
count = 0
for row in column_data:
    row_data = row.find_all('td')
    each_row_data = [data.text.strip() for data in row_data]
    print(each_row_data)
    count += 1
    if count == 10:
        break
```
