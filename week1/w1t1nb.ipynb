```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
page = requests.get("https://en.wikipedia.org/wiki/List_of_languages_by_total_number_of_speakers")
soup = BeautifulSoup(page.content, "html.parser")
```

```python
table = soup.find('table', class_ = "static-row-numbers sortable wikitable")
print(table)
```

```python
titles = table.find_all('th')
print(titles)
```

```python
headers = [title.text.strip() for title in titles]
print(headers)
```

```python
df = pd.DataFrame(columns = headers)
df
```

```python
column_data = table.find_all('tr')
for row in column_data[1:]:
    row_data = row.find_all('td')
    each_row_data = [data.text.strip() for data in row_data]
    length = len(df)
    df.loc[length] = each_row_data
```
